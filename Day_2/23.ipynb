# Comprehension


```python
names = ["Shyam", "Aniket","Bhanu","Surya","Ram",]
names2 = []
for name in names:
    if len(name) == 4:
        names2.append(name.title())
names2


```

```python
names2 = [name.title() for name in names if len(names) == 4]
```

```python
names2 = [name.title() for name in names if len(names) == 4]
names2
```

# Dict Comprehension


```python
types  = {'name': str, 'age': int, 'address': str}
```

```python
new_names = {}
for t in types:
    new_names[t] = t.title()
new_names
```

```python
new_names = {t: t.title() for t in types}
new_names
```

# Functions and Lambdas

```python
def adder(x,y):
    return x+y
adder2 = lambda x,y: x+y
adder(42, 10) == adder2(42 , 10)
```

```python
def roots(val):
    return (val**.5,-(val**.5))
roots2 = lambda val: (val**.5,-(val**.5))
roots2(4)
```

# Classes
#### .Everything is an object. You can define your own class to grp common actions with data

```python
class Duck:
    sound = "Quack Quack"
    movement = "swims in water"
    def quack(self):
        print(self.sound)
    def move(self):
        print(self.movement)
def main():
    donald = Duck()
    donald.quack()
    donald.move()
if __name__ == "__main__":
    main()
```

# Methods

```python
class Animal:
    def __init__(self,**kwargs):
        self._type = kwargs['type'] if 'type' in kwargs else 'kitten'
        self._name = kwargs['name'] if 'name' in kwargs else 'fluffy'
        self._sound = kwargs['sound'] if 'sound' in kwargs else 'meow'
        
    def type(self, t = None):
        if t:
            self._type = t
        return self._type
    def name(self, n = None):
        if n:
            self._name = n
        return self._name
    def sound(self, s = None):
        if s:
            self._sound = s
        return self._sound
    
    def __str__(self):
        return f"The {self._type} is named {self._name} and says {self._sound}"
    
def main():
    a1  = Animal(type = 'kitten', name = 'fluffy', sound = 'meow')
    a2 = Animal(type = 'duck', name = 'donald', sound = 'quack')
    print(a1)
    print(a2)

if __name__ == "__main__":
    main()  
    
```

# Inheritence

```python
class Animal:
    def type(self, t = None):
        if t: self._type = t
        try: return self._type
        except AttributeError: return None
    def name(self, n = None):
        if n: self._name = n
        try: return self._name 
        except AttributeError: return None
    def sound(self, s = None):
        if s: self._sound = s
        try: return self._sound
        except AttributeError: return None
class Duck(Animal):
    def __init__(self, **kwargs):
        self._type = 'duck'
        if 'type' in kwargs: del kwargs['type']
        super().__init__(**kwargs)
class Kitten(Animal):
    def __init__(self, **kwargs):
        self._type = 'kitten'
        if 'type' in kwargs: del kwargs['type']
        super().__init__(**kwargs)
def print_animal(o):
    if not isinstance(o, Animal):
        raise TypeError('print_animal() requires an Animal')
    print(f"The {o.type()} is named {o.name()} and says {o.sound()}")
    
```
